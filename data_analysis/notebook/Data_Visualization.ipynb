```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x, y, label='2x', color='blue', linewidth=2, marker='.', markersize=10, markeredgecolor='violet', markerfacecolor='darkblue', alpha=0.5)
plt.title('Sample Graph', fontdict={'fontname':'fira code'}, fontsize=20)
plt.xlabel('X-axis', fontdict={'fontname':'fira code'}, fontsize=20)
plt.ylabel('Y-axis', fontdict={'fontname':'fira code'}, fontsize=20)
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6])
plt.legend()
plt.savefig('mygraph.png', dpi=300)
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
values = [10, 20, 30, 40]

plt.figure(figsize=(5,3), dpi=100)
bars = plt.bar(labels, values)
pattern = ['/', 'o', '\\', '|']
for bar in bars:
    bar.set_hatch(pattern.pop(0))
# plt.savefig('hatch_graph.png', dpi=300)
plt.show()
```
