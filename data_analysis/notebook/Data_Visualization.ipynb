```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x, y, label='2x', color='blue', linewidth=2, marker='.', markersize=10, markeredgecolor='lightblue', markerfacecolor='darkblue', alpha=0.5)
plt.title('Sample Graph', fontdict={'fontname':'fira code'}, fontsize=20)
plt.xlabel('X-axis', fontdict={'fontname':'fira code'}, fontsize=20)
plt.ylabel('Y-axis', fontdict={'fontname':'fira code'}, fontsize=20)
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6])
plt.legend()
plt.savefig('mygraph.png', dpi=300)
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
values = [10, 20, 30, 40]

plt.figure(figsize=(5,3), dpi=100)
bars = plt.bar(labels, values)
pattern = ['/', 'o', '\\', '|']
for bar in bars:
    bar.set_hatch(pattern.pop(0))
# plt.savefig('hatch_graph.png', dpi=300)
plt.show()
```

```python
x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
y = [10, 15, 7, 20, 12]

plt.plot(x,y)
plt.title('Bakery')
plt.xlabel('Days')
plt.ylabel('Sales')
```

```python
months = [1,2,3,4]
sales = [1000, 1500, 1200, 1800]

plt.plot(months, sales, label='Best sales of the year', color='green', linestyle='--', linewidth=2, marker='o', markersize=10, markeredgecolor='lightgreen', markerfacecolor='darkgreen', alpha=0.5)
plt.title('Sales Data', fontdict={'fontname':'fira code'}, fontsize=20)
plt.xlabel('Months', fontdict={'fontname':'fira code'}, fontsize=20)
plt.ylabel('Sales', fontdict={'fontname':'fira code'}, fontsize=20)
plt.grid(color='gray', linestyle=':', linewidth=1)
plt.xlim(0, 5)
plt.ylim(0, 2000)
plt.xticks([1,2,3,4], ['Jan', 'Feb', 'Mar', 'Apr'])
plt.legend()
plt.show()
```

```python
products = ['A', 'B', 'C', 'D']
sales = [1000, 1500, 1200, 1800]

plt.bar(products, sales, color="purple", label='Sales') # barh for horizontal bar
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales')
plt.legend()
plt.show()
```

```python
directions = ['North', 'South', 'East', 'West']
values = [3000, 2000, 1500, 1000]
plt.pie(values, labels=directions, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'yellow'])
plt.title('Reveneu contributed by regions')
plt.show()
```

```python
scores = [80, 90, 70, 85, 95, 63, 68, 72, 88, 91]
plt.hist(scores, bins=5, color='purple', edgecolor='black')
plt.title('Distribution of Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
y = [50,55,60,65,70,75,80,85]
plt.scatter(x,y, color='red', marker='o', s=100, alpha=0.5)
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(color='gray', linestyle=':', linewidth=1)
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line Plot')
plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar Plot')
plt.suptitle('Subplots Example')
plt.tight_layout()
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

fig,ax = plt.subplots(1,2, figsize=(10,5))

ax[0].plot(x,y)
ax[0].set_title('Line Plot')

ax[1].bar(x,y)
ax[1].set_title('Bar Plot')

plt.suptitle('Subplots Example')
plt.tight_layout()
plt.show()
```
