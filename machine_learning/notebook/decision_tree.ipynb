```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
df = pd.read_csv('Dataset/data/salary.csv')
```

```python
inputs = df.drop('salary_more_then_100k', axis=1)
target = df['salary_more_then_100k']
inputs
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company'] = le_company.fit_transform(inputs['company'])
inputs['job'] = le_job.fit_transform(inputs['job'])
inputs['degree'] = le_degree.fit_transform(inputs['degree'])
inputs
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs, target)
model.score(inputs, target)
```

```python
model.predict([[2,2,1]])
```

```python
data = pd.read_csv('Dataset/data/titanic.csv')
data.head()
```

```python
df_clean = data[['Pclass', 'Sex', 'Age', 'Fare', 'Survived']].copy()
df_clean['Age'] = df_clean['Age'].fillna(df_clean['Age'].mean())
df_clean = df_clean.dropna()
inputs = df_clean[['Pclass', 'Sex', 'Age', 'Fare']]
target = df_clean['Survived']
le_sex = LabelEncoder()
inputs['Sex'] = le_sex.fit_transform(inputs['Sex'])
```

```python
model1 = tree.DecisionTreeClassifier()
model1.fit(inputs, target)
model1.score(inputs, target)
```
